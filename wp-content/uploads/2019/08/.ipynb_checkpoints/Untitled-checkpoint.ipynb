```python
import os
```

```python
for filename in os.listdir():
    #if "─▒" in filename:
        #os.rename(filename,filename.replace("─▒","i"))
    if "┼ş" in filename:
        os.rename(filename,filename.replace("┼ş","ş"))
    if "├╝" in filename:
        os.rename(filename,filename.replace("├╝","ü"))
    if "├Â─" in filename:
        os.rename(filename,filename.replace("├Â─","ö"))
    if "öşrenci" in filename:
        os.rename(filename,filename.replace("öşrenci","öğrenci"))
    if "7-sinif" in filename:
        os.rename(filename,filename.replace("7-sinif","7-sınıf"))
    if "lgs-hazirlik" in filename:
        os.rename(filename,filename.replace("lgs-hazirlik","lgs-hazırlık"))
    if "lgs-hazırlık-programi" in filename:
        os.rename(filename,filename.replace("lgs-hazırlık-programi","lgs-hazırlık-programı"))
    if "8-sinif" in filename:
        os.rename(filename,filename.replace("8-sinif","8-sınıf"))
    if "anasayfa-slideri" in filename:
        os.rename(filename,filename.replace("anasayfa-slideri","anasayfa-sliderı"))
    if "hizli-okuma-e─şitim" in filename:
        os.rename(filename,filename.replace("hizli-okuma-e─şitim","hızlı-okuma-eğitim"))
    if "anlayarak-hizli-okuma" in filename:
        os.rename(filename,filename.replace("anlayarak-hizli-okuma","anlayarak-hızlı-okuma"))
    #print(filename)
```
